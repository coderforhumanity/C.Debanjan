```python
# creating a new column for the country where matches were played
ind = ['Kolkata','Chennai','Mumbai','Mohali','Nagpur','Bengaluru','Delhi','Ahmedabad','Kanpur']
aus = ['Sydney','Adelaide','Melbourne','Brisbane']
zim = ['Harare','Bulawayo']
sri = ['Colombo (SSC)','Kandy','Galle','Colombo (RPS)']
win = ['Port of Spain','Kingston','Georgetown',"St John's",'Bridgetown']
eng = ['Nottingham',"Lord's",'The Oval','Leeds']
nz = ['Hamilton','Wellington']
pak = ['Lahore','Multan','Karachi','Rawalpindi','Faisalabad']
ban = ['Dhaka','Chattogram']
sa = ['Durban','Port Elizabeth','Johannesburg','Cape Town','Bloemfontein']
ground_country = []

for i in batting_df['Ground']:
    if i in ind:
        ground_country.append("India")
    elif i in aus:
        ground_country.append("Australia")
    elif i in zim:
        ground_country.append("Zimbabwe")
    elif i in sri:
        ground_country.append("SriLanka")
    elif i in win:
        ground_country.append("WestIndies")
    elif i in eng:
        ground_country.append("England")
    elif i in nz:
        ground_country.append("NewZealand")
    elif i in pak:
        ground_country.append("Pakistan")
    elif i in ban:
        ground_country.append("Bangladesh")
    else:
        ground_country.append("SouthAfrica")
        
batting_df['Ground Country'] = ground_country
batting_df
```

```python
# displaying the percentage of runs scored in Indian subcontinent and ouside in a pie chart
x = batting_df.groupby('Ground Country')['Runs'].sum()
indian_subcontinent = x["India"] + x["Pakistan"] + x["SriLanka"] +x["Bangladesh"]
other = x.sum()-indian_subcontinent
indian_subcontinentper = (indian_subcontinent*100)/(indian_subcontinent+other)
otherper = 100 - indian_subcontinentper

labels = "Indian Subcontinent","Outside"
sizes = [indian_subcontinentper,otherper]
explode = (0.1,0)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')
plt.title("Percentage of Runs Scored in Indian Subcontinent vs Outside")
plt.show()
```

```python
# displaying the strike rate in Indian subcontinent vs outside in data and bar chart
x = batting_df.groupby('Ground Country')['Strike Rate'].mean()
strike_rate_indian = (x["India"] + x["Pakistan"] + x["Bangladesh"] + x["SriLanka"])/4
strike_rate_outside =(x["England"]+x["NewZealand"]+x["Zimbabwe"]+x["Australia"]+x["WestIndies"]+x["SouthAfrica"])/6
y_lst = [strike_rate_indian,strike_rate_outside]
x_lst = ["Indian Subcontinent","Outside"]
sns.barplot(x = x_lst, y = y_lst)
print("Strike rate in Indian Subcontinent = ", strike_rate_indian)
print("Strike rate outside Indian Subcontinent = ", strike_rate_outside)
```

```python
# displaying the percentage of wins, loss and draw when he scores a century
x = batting_df.iloc[:,0]
y = batting_df.iloc[:,2]

win_run_lst = [y[i] for i in range(len(x)) if (x[i] == "won")]
draw_run_lst = [y[i] for i in range(len(x)) if (x[i] == "draw")]
lost_run_lst = [y[i] for i in range(len(x)) if (x[i] == "lost")]

counter1 = 0
century_lst = []

for i in win_run_lst:
    if i>=100:
        counter1 += 1
century_lst.append(counter1)
counter1 = 0
for i in draw_run_lst:
    if i>=100:
        counter1 += 1
century_lst.append(counter1)
counter1 = 0
for i in lost_run_lst:
    if i>=100:
        counter1 += 1
century_lst.append(counter1)
counter1 = 0

# applying Bayes Theorem to calculate win, draw and loss probability
win_prob = (century_lst[0]/len(win_run_lst))*(len(win_run_lst)/100)/(sum(century_lst)/100)*100
draw_prob = (century_lst[1]/len(win_run_lst))*(len(win_run_lst)/100)/(sum(century_lst)/100)*100
lost_prob = (century_lst[2]/len(win_run_lst))*(len(win_run_lst)/100)/(sum(century_lst)/100)*100

labels = "Win","Draw","Lost"
sizes = [win_prob,draw_prob,lost_prob]
explode = (0.1,0.1,0)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')
plt.title("Percentage of Wins, Draws and Losses when he scores a century")
plt.show()
```
