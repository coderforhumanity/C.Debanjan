```python
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier, XGBRFClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score
import joblib
```

```python
df = pd.read_csv("../Data/credit_risk_dataset_prepared.csv")
df
```

```python
x = df.drop(labels = ["loan_status"], axis=1)
y = df["loan_status"]
```

```python
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=7)
fold_no = 1
for train_index, test_index in skf.split(x,y):
    train = df.loc[train_index,:]
    test = df.loc[test_index,:]
    train_filename = 'train_split_' + str(fold_no) + '.csv'
    test_filename = 'test_split_' + str(fold_no) + '.csv'
    train.to_csv('../Data/splits/' + train_filename, index=False)
    test.to_csv('../Data/splits/' + test_filename, index=False)
    fold_no += 1
```

```python
train_split_1 = pd.read_csv("../Data/splits/train_split_1.csv")
test_split_1 = pd.read_csv("../Data/splits/test_split_1.csv")
xtrain = train_split_1.drop(labels = ["loan_status"], axis=1)
ytrain = train_split_1["loan_status"]
scaler = StandardScaler()
train_split_1_scaled = scaler.fit_transform(xtrain)
logistic_regression = LogisticRegression()
grid={"C":np.logspace(-3,3,4), "penalty":["l1","l2","elasticnet","none"], "solver":['newton-cg', 'lbfgs'], "max_iter":[100,500,1000]}
logreg_cv=GridSearchCV(logistic_regression,grid,cv=10, verbose=3)
logreg_cv.fit(xtrain, ytrain)
print("tuned hpyerparameters :(best parameters) ",logreg_cv.best_params_)
```

```python
logistic_regression = LogisticRegression(C=0.001, max_iter=500, penalty='none', solver='lbfgs')
xtrain = train_split_1.drop(labels = ["loan_status"], axis=1)
ytrain = train_split_1["loan_status"]
xtest = test_split_1.drop(labels = ["loan_status"], axis=1)
ytest = test_split_1["loan_status"]
scaler = StandardScaler()
train_split_1_scaled = scaler.fit_transform(xtrain) 
logistic_regression.fit(xtrain, ytrain)
ypred = logistic_regression.predict(xtest)
print(confusion_matrix(ytest, ypred))
print(classification_report(ytest, ypred))
auc = roc_auc_score(ytest, ypred)
print(auc)
```
